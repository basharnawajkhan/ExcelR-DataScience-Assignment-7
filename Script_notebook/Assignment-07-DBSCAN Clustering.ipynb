# 1) DBSCAN Clustering - crime_data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler as std
```

```python
crime = pd.read_csv("C:/Users/BASHAR/OneDrive/Documents/Data Science/Assignment/Assignment - 7/crime_data.csv")
crime.head()
```

## EDA

```python
crime.info()
```

```python
crime.isna().sum()
```

```python
crime[crime.duplicated()].shape
```

```python
crime1 = crime.drop(["Unnamed: 0"], axis=1)
crime1.head()
```

## Standardization

```python
STD = std().fit_transform(crime1)
crime_std = pd.DataFrame(STD, columns=crime1.columns)
crime_std.head()
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(crime_std)
```

```python
dbscan.labels_
```

```python
Clust = pd.DataFrame(dbscan.labels_, columns=["Cluster"])
Clust.head()
```

```python
pd.concat([crime, Clust], axis=1).head()
```

#### Alternate way

```python
crime["Cluster"] = dbscan.labels_
crime.head()
```

```python
Cluster_A = crime.loc[crime["Cluster"] == 0]
Cluster_A
```

```python
Cluster_B = crime.loc[crime["Cluster"] == -1]
Cluster_B
```

```python
Cluster_C = crime.loc[crime["Cluster"] == 1]
Cluster_C
```

```python
round(crime.groupby("Cluster").mean(),2)
```

# 2) DBSCAN Clustering - EastWestAirlines

```python
air = pd.read_csv("C:/Users/BASHAR/OneDrive/Documents/Data Science/Assignment/Assignment - 7/EastWestAirlines.csv")
air.head()
```

## EDA

```python
air1 = air.drop(["ID#"], axis=1)
air1.head()
```

## Standardization
